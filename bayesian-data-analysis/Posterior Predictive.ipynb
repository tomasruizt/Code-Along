```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

# Plotting functions
Below are helper functions to plot beta and binomial distributions.

```python
def plot_beta_hist(a: float, b: float) -> None:
    xs = np.linspace(0, 1)
    ys = scipy.stats.beta.pdf(xs, a=a, b=b)
    plt.plot(xs, ys)
    plt.fill_between(xs, 0, ys, alpha=0.2)
    plt.xlim(-0.2, 1.2)
    
n = 100
def plot_binomial_hist(p: float) -> None:
    xs = np.arange(n)
    ys = scipy.stats.binom.pmf(xs, n=n, p=p)
    plt.scatter(xs, ys)

def plot_hist(xs) -> None:
    plt.hist(xs, bins=20, density=True)
    plt.show()
```

```python
plot_beta_hist(a=3, b=1)
```

```python
plot_binomial_hist(p=0.5)
```

# Sampling from the posterior predictive
* Assume the Likelihood is a binomial distribution with $n$ trials, and unkown probability of success $y$:
$$p(s|y) = B(n, y)$$

* The prior for $y$ is uniformly distributed within $[0, 1]$. We don't assume to know anything about it.
* This can be modeled with a beta distribution with parameters $\theta = (a, b) = (1, 1)$

$$p(y|\theta) = Beta(\theta) = Beta(a, b)$$



```python
num_samples = 12345
ys = np.random.beta(5, 10, size=num_samples)
ss = np.random.binomial(n=n, p=ys, size=num_samples)
plt.hist(ss, bins=100)
plt.show()
```
