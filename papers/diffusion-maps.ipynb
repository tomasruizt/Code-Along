```python
import numpy as np
import matplotlib.pyplot as plt
```

# Diffusion Map
Example from Wikipedia: https://en.wikipedia.org/wiki/Diffusion_map

```python
t = np.linspace(0, 2 * np.pi, 1000)
x = np.sin(t)
y = np.cos(t)
```

```python
n = len(x)
plt.scatter(x[:n], y[:n])
plt.grid()
```

```python
x2 = 0.5 * np.sin(8 * t)
y2 = 0.5 * np.cos(8 * t)
```

```python
X = x * (1 + x2)
Y = y * (1 + x2)
plt.scatter(X, Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X, Y, y2)
```
